```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import mglearn

%matplotlib inline
```

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

irisDataset = load_iris()
```

```python
X_train, X_test, y_train, y_test = train_test_split(irisDataset['data'], irisDataset['target'], random_state=0)
```

```python
irisDataframe = pd.DataFrame(X_train, columns=irisDataset.feature_names)
grr = pd.scatter_matrix(irisDataframe, c=y_train, figsize=(15, 15), marker='o', hist_kwds={'bins': 20}, s=60, alpha=.8, cmap=mglearn.cm3)
```

```python
knn = KNeighborsClassifier(n_neighbors=1)

knn.fit(X_train, y_train)
```

```python
print(knn.score(X_test, y_test))
```
