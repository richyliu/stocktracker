```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
import pandas
import matplotlib

%matplotlib inline

irisDataset = load_iris()
```

```python
X_train, X_test, y_train, y_test = train_test_split(irisDataset['data'], irisDataset['target'], random_state=0)
```

```python
irisDataframe = pandas.DataFrame(X_train, columns=irisDataset.feature_names)
grr = pandas.scatter_matrix(irisDataframe, c=y_train, figsize=(15, 15), marker='o', hist_kwds={'bins': 20}, s=60, alpha=.8, cmap=matplotlib.colors.ListedColormap(['#0000aa', '#ff2020', '#50ff50']))
```
